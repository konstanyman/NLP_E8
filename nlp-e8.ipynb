```python
# imports
from conllu import parse
```

```python
# read data
with open("en_ewt-ud-train.conllu", "r", encoding="utf-8") as f:
    data = f.read()
```

```python
# parse data
sentences = parse(data)
```

```python
i = 0
for sentence in sentences:
    for token in sentence:
        print(token)
        i += 1
print(i)
```
